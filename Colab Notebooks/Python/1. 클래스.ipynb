#### __init__(self)
+ 생성자, 클래스 인스턴스가 생성될 때 호출됨
+ self 인자는 항상 첫번째에 오며 자기 자신을 가리킴
+ 이름이 꼭 self일 필요는 없지만, 관례적으로 self로 사용

+ 생성자에서는 해당 클래스가 다루는 데이터를 정의
  - 이 데이터를 멤버변수(member variable) 또는 속성(attribute)라고 함


#### **self**
 - 파이썬의 method는 항상 첫번째 인자로 self를 전달
 - self는 현재 해당 메서드가 호출되는 객체 자신을 가리킴
 - C++/C#, Java 의 this에 해당
 - 이름이 self일 필요는 없으나, 위치는 항상 맨 처음의 parameter이며 관례적으로 self로 사용

#### **special method**

#### 클래스 상속 (Class inheritance)

- Notebook들은 제조사마다 다른 모델이 있습니다.
- 해당 모델들은 NoteBook 이지만, 모델마다 다른 특징을 가지고 있습니다.
- 노트북의 개념을 그대로 이어받아, 모델마다 다른 개념을 CLass로 구현해 보시오.

```python
# notebook class를 상속받아서 새로운 MacBook class를 정의
class MacBook(Notebook):
  def __init__(self, manufacture, model, release_date, displat_sizs, cpu, ram, ssd, os):
    self.manufacture = manufacture
    self.model = model
    self.release_date = release_date
    self.displat_sizs = displat_sizs
    self.cpu = cpu
    self.ram = ram
    self.ssd = ssd
    self.os = os

macbook = MacBook("Apple", "Macbook Pro 16", "2022","16","M1 Max", 64, 2048, "macOS")
macbook.print_model()
print(macbook.is_UNIX())
```

```python
# notebook class 를 상속받아서 새로운 Lenovo class를 정의
# yoga 속성
class Lenovo(Notebook):
  def __init__(self, manufacture,model, yoga, displat_sizs, cpu, ram, ssd, os="Windows 11"):
    self.manufacture = manufacture
    self.model = model
    self.yoga = yoga
    self.displat_sizs = displat_sizs
    self.cpu = cpu
    self.ram = ram
    self.ssd = ssd
    self.os = os

lenovo = Lenovo("Lenovo","yoga slim7 13.3","14 Gen 5","16","AMD Ryzen", 64, 2048)
lenovo.print_model()
print(lenovo.is_UNIX())
```
