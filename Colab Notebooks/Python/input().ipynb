# 윤년 계산하기
 - 사용자 입력을 받을 숫자 num이 있습니다.
 - 주어진 숫자 num이 윤년인지 아닌지 출력하는 코드를 작성하시오.

- 윤년 규칙
 1. 4의 배수일 것
 2. 100의 배수가 아닐것 (1번 조건 무시)
 3. 400의 배수일 것

```python
num = int(input("년도를 입력하세요 : "))
if num%4==0 and num%100!=0 or num%400==0:
  print("윤년")
else:
  print("평년")
```
