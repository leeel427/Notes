```python
import torch
```

```python
from torch._C import device
print(torch.tensor([[1, 2],[3, 4]], device="cuda:0"))
```

```python
data= [
    [1, 2],
    [3, 4]
]

x = torch.tensor(data)
print(x.is_cuda)          # GPU 사용 여부 체크하기
```

```python
# GPU로 옮기기
x = x.cuda()
print(x.is_cuda)
```

```python
# CPU로 옮기기
x = x.cpu()
print(x.is_cuda)
```

```python
# GPU 장치의 텐서
a = torch.tensor([
    [1, 1],
    [2, 2]
]).cuda()

# CPU 장치의 텐서
b = torch.tensor([
    [5, 6],
    [7, 8]
])

#print(torch.matmul(a, b))     # 오류 발생
print(torch.matmul(a.cpu(), b))
```
